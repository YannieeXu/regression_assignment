```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("../manual/regression_data.csv")
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df["YearsExperience"], df["Salary"], color='red')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
X = df[["YearsExperience"]]
y = df["Salary"]
model = LinearRegression().fit(X, y)
y_pred = model.predict(X)
print("Slope: %.4f" % model.coef_[0])
print("Intercept: %.4f" % model.intercept_)
print("R-squared: %.4f" % r2_score(y, y_pred))
```

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y, label="Data", color='red')
plt.plot(X, y_pred, color="blue", lw=2, label="Regression Line")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Linear Regression (AI Version)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
