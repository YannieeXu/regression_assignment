# Linear Regression with Diagnostics (R - AI Generated)

This notebook fits a linear regression model and computes diagnostics.

```r
library(ggplot2)
data <- read.csv("regression_data.csv")
head(data)
```

## Diagnostics

Compute slope, intercept, Pearson r, and MSE.

```r
x <- data$YearsExperience
y <- data$Salary
model <- lm(y ~ x)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r_val <- cor(x, y)
pred <- predict(model)
mse <- mean((y - pred)^2)
cat("Slope:", round(slope, 4), "\n")
cat("Intercept:", round(intercept, 4), "\n")
cat("r:", round(r_val, 4), "\n")
cat("MSE:", round(mse, 4), "\n")
```

```r
p <- ggplot(data, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red", size = 2) +
  geom_smooth(method = "lm", se = FALSE, color = "blue", linewidth = 1) +
  annotate("text", x = min(x), y = max(y),
           label = sprintf("y = %.2fx + %.2f\nr = %.2f\nMSE = %.0f", slope, intercept, r_val, mse),
           hjust = 0, vjust = 1, size = 4) +
  labs(title = "Linear Regression with Diagnostics", x = "YearsExperience", y = "Salary") +
  theme_bw()
p
```
