```r
df <- read.csv("../manual/regression_data.csv")
head(df)
```

```r
plot(df$YearsExperience, df$Salary, main="Salary vs Experience (AI Version)", xlab="YearsExperience", ylab="Salary", pch=19, col="steelblue")
```

```r
model <- lm(Salary ~ YearsExperience, data=df)
summary(model)
```

```r
plot(df$YearsExperience, df$Salary, main="Salary vs Experience (AI Version)", xlab="YearsExperience", ylab="Salary", pch=19, col="steelblue")
abline(model, col="red", lwd=2)
cat(sprintf("R-squared: %.4f\n", summary(model)$r.squared))
```
