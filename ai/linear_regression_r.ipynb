```r
df <- read.csv("../manual/regression_data.csv")
head(df)
```

```r
plot(df$x, df$y, main="Scatter Plot (AI Version)", xlab="X", ylab="Y", pch=19, col="steelblue")
```

```r
model <- lm(y ~ x, data=df)
summary(model)
```

```r
plot(df$x, df$y, main="Linear Regression (AI Version)", xlab="X", ylab="Y", pch=19, col="steelblue")
abline(model, col="red", lwd=2)
cat(sprintf("R-squared: %.4f\n", summary(model)$r.squared))
```
