# Linear Regression Analysis (R)

This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.


```r
df <- read.csv('manual/regression_data.csv')
head(df)
```

```r
plot(df$YearsExperience, df$Salary, main='Salary vs Experience', xlab='YearsExperience', ylab='Salary', pch=19, col='blue')
```

## Model Fitting

Now we fit a linear regression model to examine the relationship between Years of Experience and Salary.


```r
model <- lm(Salary ~ YearsExperience, data=df)
summary(model)
```

```r
plot(df$YearsExperience, df$Salary, main='Salary vs Experience', xlab='YearsExperience', ylab='Salary', pch=19, col='blue')
abline(model, col='red', lwd=2)
r2 <- summary(model)$r.squared
legend('topleft', legend=c('Data', 'Regression Line'), col=c('blue', 'red'), pch=c(19, NA), lty=c(NA, 1))
text(0.5, 30, sprintf('R-squared: %.4f', r2), adj=0)
```

## Results

The regression line shows a positive relationship between experience and salary. The R-squared value indicates how well the model fits the data.
