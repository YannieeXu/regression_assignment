# Linear Regression Analysis (R)

This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.


```r
library(ggplot2)
df <- read.csv('regression_data.csv')
head(df)
```

```r
plot(df$YearsExperience, df$Salary, main='Salary vs Experience', xlab='Years of experience', ylab='Salary', pch=19, col='blue')
```

## Model Fitting

Now we fit a linear regression model to examine the relationship between Years of Experience and Salary.


```r
model <- lm(Salary ~ YearsExperience, data=df)
summary(model)
```

```r
ggplot() +
  geom_point(aes(x = df$YearsExperience, y = df$Salary), colour = 'red') +
  geom_line(aes(x = df$YearsExperience, y = predict(model, newdata = df)), colour = 'blue') +
  ggtitle('Salary vs Experience') +
  xlab('Years of experience') +
  ylab('Salary')
cat("R-squared:", round(summary(model)$r.squared, 4), "\n")
```

## Results

The regression line shows a positive relationship between experience and salary. The R-squared value indicates how well the model fits the data.
