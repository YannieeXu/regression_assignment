# Linear Regression Analysis (Python)

This notebook demonstrates a simple linear regression analysis using Python to model Salary based on Years of Experience.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv("regression_data.csv")
data.head()
```

## Model Fitting and Diagnostics

We fit a linear regression model and calculate diagnostics: slope, intercept, correlation coefficient, and MSE.

```python
x = data["YearsExperience"]
y = data["Salary"]
slope, intercept, r_value, p_value, std_err = linregress(x, y)
y_pred = slope * x + intercept
mse = mean_squared_error(y, y_pred)
print(f"Slope: {slope:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"r: {r_value:.4f}")
print(f"MSE: {mse:.4f}")
```

```python
plt.scatter(x, y, color="red")
plt.plot(x, y_pred, color="blue")
plt.text(0.05, 0.95, f"y = {slope:.2f}x + {intercept:.2f}" + f"
r = {r_value:.2f}" + f"
MSE = {mse:.2f}", transform=plt.gca().transAxes, fontsize=10, verticalalignment="top")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("Salary vs YearsExperience")
plt.show()
```

## Results

The regression line shows a positive relationship. The diagnostics help evaluate model fit.
