# Linear Regression Analysis (R)

This notebook demonstrates a simple linear regression analysis using R to model Salary based on Years of Experience.

```r
library(ggplot2)
data <- read.csv("regression_data.csv")
head(data)
```

## Model Fitting and Diagnostics

We fit a linear regression model and calculate diagnostics: slope, intercept, correlation coefficient, and MSE.

```r
x <- data$YearsExperience
y <- data$Salary
model <- lm(y ~ x)
slope <- coef(model)[2]
intercept <- coef(model)[1]
r_value <- cor(x, y)
pred <- predict(model)
mse <- mean((y - pred)^2)
cat("Slope:", round(slope, 4), "\n")
cat("Intercept:", round(intercept, 4), "\n")
cat("r:", round(r_value, 4), "\n")
cat("MSE:", round(mse, 4), "\n")
```

```r
p <- ggplot(data, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red") +
  geom_smooth(method = "lm", se = FALSE, color = "blue") +
  annotate("text", x = min(x) + 0.1 * (max(x) - min(x)), y = max(y) * 0.95,
           label = paste("y =", round(slope, 2), "x +", round(intercept, 2),
                         "\nr =", round(r_value, 2), "\nMSE =", round(mse, 2)),
           hjust = 0, size = 4) +
  labs(title = "Salary vs YearsExperience",
       x = "Years of Experience", y = "Salary") +
  theme_minimal()
p
```

## Results

The regression line shows a positive relationship. The diagnostics help evaluate model fit.
