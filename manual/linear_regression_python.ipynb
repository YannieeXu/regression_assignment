# Linear Regression Analysis (Python)

This notebook demonstrates a simple linear regression analysis using Python to model Salary based on Years of Experience.


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('regression_data.csv')
dataset.head()
```

```python
plt.scatter(dataset['YearsExperience'], dataset['Salary'], color='red')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
```

## Model Fitting

Now we fit a linear regression model to examine the relationship between Years of Experience and Salary.


```python
X = dataset[['YearsExperience']]
y = dataset['Salary']
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print(f'Slope: {model.coef_[0]:.4f}')
print(f'Intercept: {model.intercept_:.4f}')
print(f'R-squared: {r2_score(y, y_pred):.4f}')
```

```python
plt.scatter(dataset['YearsExperience'], dataset['Salary'])
plt.plot(dataset['YearsExperience'], y_pred, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Salary vs Experience')
```

## Results

The regression line shows a positive relationship between experience and salary. The R-squared value indicates how well the model fits the data.
